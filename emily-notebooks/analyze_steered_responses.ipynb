## Analyze Giant Spreadsheet

```python
massive_steered_responses = pd.read_csv(f"{RESPONSES_DIR}/vsorted_llama3_massive_steered_responses.csv", index_col=[0,1])

interesting_responses = pd.read_csv(f"{RESPONSES_DIR}/interesting_llama3_responses.csv")
interesting_responses = interesting_responses.drop(columns=["Unnamed: 1", "Unnamed: 2", "TODO add ones with kids"])
```

```python
interesting_responses_stats = interesting_responses.groupby("run_name").count()[
    ["admits", "refusal", "ai", "im_john_cheating", "half_admits"]
].reset_index().loc[1:]

interesting_responses_stats = interesting_responses_stats.sort_values("run_name", key=lambda col: col.str.split("_", expand=True)[1] + col)
```

```python
interesting_responses_stats[["ortho", "norm", "pq", "t"]] = interesting_responses_stats["run_name"].str.split("_", expand=True)
interesting_responses_stats["ortho"] = interesting_responses_stats["ortho"].str.removeprefix("o")
interesting_responses_stats["norm"] = interesting_responses_stats["norm"].str.removeprefix("n").astype(float)
interesting_responses_stats["p"] = interesting_responses_stats["pq"].str.get(1).astype(int)
interesting_responses_stats["q"] = interesting_responses_stats["pq"].str.get(-1).astype(int)
interesting_responses_stats["q==p"] = interesting_responses_stats["pq"].str.get(-1).astype(int) == 1
```

```python
interesting_responses_stats
```

```python
norm_stats_df = interesting_responses_stats[["norm", "admits", "refusal", "ai", "im_john_cheating", "half_admits"]].groupby("norm").sum()
norm_stats_df.plot.bar()
```

```python
group_bys = ["norm", "p", "q==p"]
for gb in group_bys[:1]:
    stats_df = interesting_responses_stats[[gb, "admits", "refusal", "ai"]].groupby(gb).sum()
    stats_df.plot.bar(xlabel=f"Hyperparameter: Magnitude of Steering Vector", ylabel="Response Type Count", title=f"Type of Response by Hyperparameter: Magnitude of Steering Vector")
```
